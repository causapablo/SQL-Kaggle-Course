# Due to we have a limit size of data we can retrieve per month we can check how much data we 

```python
query = """
        SELECT score, title
        FROM `bigquery-public-data.hacker_news.full`
        WHERE type = "job"
        """
```

```python
dry_run_config = bigquery.QueryJobConfig(dry_run=True)
dry_run_query_job = client.query(query, job_config = dry_run_config)
```

```python
print("This query will process {} bytes".format(dry_run_query_job.total_bytes_processed))
```

**Also, we can limit how much data we retrieve in a query**

```python
ONE_HUNDRED_MB=1000*1000*1000
safe_config = bigquery.QueryJobConfig(maximum_bytes_billed=ONE_HUNDRED_MB)
safe_query_job = client.query(query,job_config = safe_config)
```

```python
safe_query_job.to_dataframe()
```

**Query for the question  "Which countries have reported levels in units of "ppm""?**

```python
query = """
        SELECT country
        FROM `bigquery-public-data.openaq.global_air_quality`
        WHERE unit = 'ppm'
        """
```

```python
safe_config = bigquery.QueryJobConfig(maximum_bytes_billed = 10**10)
safe_query_job = client.query(query, job_config = safe_config)
```

```python
first_results = safe_query_job.to_dataframe()
```

```python
print(first_results.head())
```

**Which pollution levels were reported to be exactly 0?**

```python
zero_pollution_query = """
                       SELECT *
                       FROM `bigquery-public-data.openaq.global_air_quality`
                       WHERE value = 0
                       """
```

```python
safe_config = bigquery.QueryJobConfig(maximum_bytes_billed = 10**10)
job_query = client.query(zero_pollution_query,job_config = safe_config)
```

```python
zero_pollution_results = job_query.to_dataframe()
```
