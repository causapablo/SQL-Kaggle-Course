# Practice on Hacker News Dataset.
**What are the authors with more than 10000 posts?**

```python
prolific_commeters_query = """
                           SELECT author, COUNT(id) AS NumPosts
                           FROM `bigquery-public-data.hacker_news.comments`
                           GROUP BY author
                           HAVING COUNT(id) > 10000
                           """
```

```python
safe_config = bigquery.QueryJobConfig(maximum_bytes_billed = 10**10)
query_job = client.query(prolific_commeters_query, job_config = safe_config)
```

```python
prolific_commeters = query_job.to_dataframe()
prolific_commeters.head()
```

# ORDER BY and EXTRACT()
**Which day of the week has the most fatal motor accidents?**

```python
dataset_ref = client.dataset("nhtsa_traffic_fatalities", project = "bigquery-public-data")
dataset = client.get_dataset(dataset_ref)
```

```python
tables = list(client.list_tables(dataset))
for table in tables:
    print(table.table_id)
```

```python
table_ref = dataset_ref.table('accident_2015')
table = client.get_table(table_ref)
```

```python
table.schema
```

```python
client.list_rows(table, max_results = 5).to_dataframe()
```

```python
query = """
        SELECT EXTRACT(DAY from timestamp_of_crash) AS day_of_week, COUNT(consecutive_number) AS num_accidents
        FROM `bigquery-public-data.nhtsa_traffic_fatalities.accident_2015`
        GROUP BY day_of_week
        ORDER BY num_accidents DESC
        """
```

```python
safe_config = bigquery.QueryJobConfig(maximum_bytes_biller = 10**10)
query_job = client.query(query,job_config = safe_config)
```

```python
accidents = query_job.to_dataframe()
accidents.head()
```

```python
query_new = """
        SELECT EXTRACT(DAYOFWEEK FROM timestamp_of_crash) AS day_of_week, COUNT(1) AS num_accidents
        FROM `bigquery-public-data.nhtsa_traffic_fatalities.accident_2015`
        GROUP BY day_of_week
        ORDER BY num_accidents DESC
        """
```

```python
safe_config = bigquery.QueryJobConfig(maximum_bytes_biller = 10**9)
query_job = client.query(query_new,job_config = safe_config)
```

```python
accidents_new = query_job.to_dataframe()
accidents_new.head()
```

```python
query_third = """
        SELECT EXTRACT(DAYOFWEEK FROM timestamp_of_crash) AS day_of_week,
               COUNT(day_of_week) AS num_accidents
        FROM `bigquery-public-data.nhtsa_traffic_fatalities.accident_2015`
        GROUP BY day_of_week
        ORDER BY num_accidents DESC
        """
```

```python
safe_config = bigquery.QueryJobConfig(maximum_bytes_biller = 10**9)
query_job = client.query(query_third,job_config = safe_config)
```

```python
accidents_third = query_job.to_dataframe()
accidents_third.head()
```

```python
query = """
        SELECT COUNT(consecutive_number) AS num_accidents, 
               EXTRACT(DAYOFWEEK FROM timestamp_of_crash) AS day_of_week
        FROM `bigquery-public-data.nhtsa_traffic_fatalities.accident_2015`
        GROUP BY day_of_week
        ORDER BY num_accidents DESC
        """
```

```python
# Set up the query (cancel the query if it would use too much of 
# your quota, with the limit set to 1 GB)
safe_config = bigquery.QueryJobConfig(maximum_bytes_billed=10**9)
query_job = client.query(query, job_config=safe_config)

# API request - run the query, and convert the results to a pandas DataFrame
accidents_by_day = query_job.to_dataframe()

# Print the DataFrame
accidents_by_day
```

```python
#Conclusions:
# 1) No matter what we use in COUNT() as argument, it can be "1", "day_of_week" or "consecutive_number"
# the task will be performed is to count all the instances in each group of "day_of_week"
# 2) There's a difference between the DataFrame given by the value DAY and DAYOFWEEK
# Because DAY is the day in the month, from 1 to 31, and DAYOFWEEK is the day of week, from 1 to 7.
```
